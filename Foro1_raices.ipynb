# Cálculo de la raíz por el método de la bisección


```python

import numpy as np
def biseccion(f,a,b,tol,nmax):
    for i in range(1,nmax):
        if np.sign(f(a)) == np.sign(f(b)):
            print("La raíz no está en el intervalo especificado")
            return
        
        m = (a + b)/2
        
        if abs(f(m)) < tol:
            return m,i
        elif np.sign(f(a)) == np.sign(f(m)):
            a = m
        else:
            b = m
        i += i
```

```python

import math
f = lambda x: math.log(x+1e-15) + x # Añado un valor infinitesimal para evitar que se calcule el ln(0) que sabemos que no existe
r1,cont = biseccion(f, 0, 1, 0.000001,100)
print("Con ",cont," iteraciones, la raiz es ",r1," y el error es ",0.575-r1) # 0.575 será el valor aproximado de la raiz
```
